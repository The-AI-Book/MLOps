### ML Metadata library

- Tracks metadata flowing between components in pipeline
- Supports multiple storage backends

### ML Metadata terminology

**Units**: 
- Artifact: is an elementary unit of data that gets fed into the ML metadata store and as the data is consumed as input or generated as output of each component 
- Execution: is a record of any component run during the ML pipeline workflow, along with its associated runtime parameters
- Context: may hold the metadata of the projects beign run, experiments beign conducted, details about pipelines, etc.
