## Outline
---
- Overview of advanced labeling techniques
- **Semi-supervised learning:**  How can we assign labels in other ways other than going through each example manually? In other words, can we automate the process even at the expense of maybe introducing inaccuracies in the labeling process?
- **Active learning:** It uses intelligence sampling to assign labels based on the existing data to unlabeled data
- **Weak supervision:** is a way to programmatically labeled data, typically by going or using heuristics which are designed by subject matter experts

## Why is advanced labeling important?

- ML use is growing everywhere as is the need for labeled training sets

## Semi-supervised Learning
---

- Small pool human labeled data
- Large pool unlabeled data

$\Longrightarrow$ Train your model.

It relies on some degree of uniformity or clustering within feature space

### Human labeling, Semi-supervised

Advantages:
- Combining labeled and unlabeled data boosts accuracy
- Getting unlabeled data is cheap

### Label propagation
- Semi-supervised ML algorithm
- A subset of the examples have labels
- Labels are propagated to the unlabeled points: Based on similarity or "community structure"

## Active Learning

---

- A family of algorithms for intelligently sampling data
- Select the points tobe labeled that would be most informative for model training

Very helpful in the following situations:

- Contrained data budgets: you can only afford labeling a few points
- Imbalanced dataset: helps selecting rare classes for training
- Target metrics: when baseline sampling strategy does not improve selected metrics

### Active learning sampling techniques

- **Margin sampling:** Label points the current model is least confident in
- **Cluster-base sampling:** Sample from well-formed clusters to "cover" the entire spaces
- **Query-by-committe:** train an ensemble of models and sample points that generate disagreement
- **Region-based sampling:** Runs several active learning algorithms in different partitions of the space

## Weak Supervision

---

- Weak supervision is a way to generate labels using information from one or more sources. Usualley these are subject matter experts, and usually they're designing heuristics. More specifically weak supervision comprises one or more noisy condicional distributions over unlabeled data.

- The main objective is to learn a generative model that determines the relevance of each of these noisy sources.
