## Search Strategies

---

A few search strategies: 
1. Grid Search
2. Random Search
3. Bayesian Optimization
4. Evolutionary Algorithms
5. Reinforcement Learning

### Grid & Random Search
- Grid Search: Exhaustive search approach on fixed grid values
- Random Search.

Grid Search and Random Search are both suited for smaller search spaces. Both quickly fail with growing size of search space.

### Bayesian Optimization
- Assumes that a specific probability distribution, is underlying the performance
- Tested architecture constrain the probability distribution and guide the selection of the next option

## Measuaring AutoML Efficacy
---
- Validation Accuracy: Computationally heavy, expensive
Strategies to Reduce the Cost
1. Lower fidelity estimates
2. Learning Curve Extrapolation
3. Weight Inheritance / Network Morphisms


### Lower fidelity estimates
- Data subset
- Low resolution images
- Fewer filters and cells
- It reduce cost but underestimates performance

### Learning Curve Extrapolation
- Requirees predicting the learning curve reliably
- Extrapolates based on initial learning
- Removes poor performers

### Weight Inheritance / Networks Morphisms
- Initialize weights of new architectures based on previosly trained architectures (similar to transfer learning)
- Uses Network Morphism
Underlying function unchanged:
- New network inherist knowledge from parent network
- Computational speed up: only a few days of GPU usage
